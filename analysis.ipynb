```python
import pandas as pd

# Load the CSV file
file_path = '/Users/subash/Desktop/datasets/archive/wisdm-dataset/wisdm-dataset/arff.csv'
data = pd.read_csv(file_path)

# Display basic information about the dataset
data_info = data.info()
data_head = data.head()

data_info, data_head
```

```python
# Check for missing values
missing_values = data.isnull().sum()

# Summarize basic statistics
data_description = data.describe()

missing_values, data_description
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the plotting style
sns.set(style="whitegrid")

# Plot the distribution of activities
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='ACTIVITY', order=data['ACTIVITY'].value_counts().index)
plt.title('Distribution of Activities')
plt.xlabel('Activity')
plt.ylabel('Count')
plt.show()


```

```python
# Plot the distribution of subject IDs (class)
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='class', order=data['class'].value_counts().index)
plt.title('Distribution of Subject IDs')
plt.xlabel('Subject ID')
plt.ylabel('Count')
plt.show()

```

```python
# Correlation heatmap of a subset of features (to avoid clutter)
subset = data[['XAVG', 'YAVG', 'ZAVG', 'XPEAK', 'YPEAK', 'ZPEAK', 'XABSOLDEV', 'YABSOLDEV', 'ZABSOLDEV', 'RESULTANT']]
plt.figure(figsize=(12, 10))
correlation_matrix = subset.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.show()

```

```python
# Plot distributions of key features
features_to_plot = ['XAVG', 'YAVG', 'ZAVG', 'XPEAK', 'YPEAK', 'ZPEAK', 'RESULTANT']
plt.figure(figsize=(15, 10))
for i, feature in enumerate(features_to_plot, 1):
    plt.subplot(3, 3, i)
    sns.histplot(data[feature], kde=True)
    plt.title(f'Distribution of {feature}')
plt.tight_layout()
plt.show()
```
